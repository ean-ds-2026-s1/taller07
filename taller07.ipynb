## Equipos de Fútbol

Vamos a desarrollar una serie de operaciones sobre una lista de nodos donde cada uno de estos
contiene la información de lo realizado por un equipo de fútbol de la Liga Española. 

Cada equipo tiene los atributos y métodos que se muestran en el siguiente diagrama de clases:

![Equipo](https://raw.githubusercontent.com/luiscobo/poo/refs/heads/main/EquipoDeFutbol.png)

Donde tenemos la siguiente descripción de los atributos de la clase:

* `año`: temporada o año del cual tenemos la información del equipo.
* `nombre`: el nombre del equipo de fútbol.
* `partidos_ganados_local`: el número de partidos que ganó el equipo en el año dado jugando de local.
* `partidos_ganados_visitante`: el número de partidos que ganó el equipo de visitantes
* `partidos_perdidos_local`: el número de partidos que perdió el equipo en el año dado jugando de local.
* `partidos_perdidos_visitante`: el número de partidos que perdió el equipo jugando de visitante.
* `partidos_empatados`: el número total de partidos que empató el equipo en todo el año.
* `goles_local`: número total de goles que hizo el equipo jugando de local
* `goles_visitante`: el número total de goles que hizo el equipo en el año jugando de visitante
* `goles_en_contra`: el número total de goles que recibió el equipo en todo el año.

La clase, además de los atributos anteriores, posee los siguientes métodos:

* `partidos_ganados()`: Método que no recibe parámetros y que retorna el total de todos los partidos que
   ganó el equipo, teniendo en cuenta los ganados como local y como visitante.
* `partidos_perdidos()`: Método que no recibe parámetros y que retorna el total de todos los partidos que
  haya perdido el equipo en el año.
* `partidos_jugados()`: Métodos que no recibe parámetros y que retorna el total de partidos jugados por el
   equipo, teniendo en cuenta los partidos ganados, empatados y perdidos.
* `puntos()`: Método que no recibe parámetros y que retorna el total de puntos que hizo el equipo en el año
  en cuestión, teniendo en cuenta que cada partido ganado tiene 3 puntos y cada partido empatado tiene 1 punto.
* `goles_a_favor()`: Método que no recibe parámetros y que retorna la cantidad de goles que hizo el equipo 
  teniendo en cuenta los goles hecho de local y los goles hechos de visitante.
* `diferencia_de_goles()`: Método que no recibe parámetros y que retorna la diferencia entre los goles hechos
  por el equipo (los goles a favor) y los goles que recibió el equipo (los goles en contra).
